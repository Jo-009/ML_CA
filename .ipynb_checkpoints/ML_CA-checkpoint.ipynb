#### Creating X_train

```python
x_train = []
for label in y_labels:
    if label[-3:] == 'xml':
        continue
    else:
        x_train.append(label)
```

#### Creating y_train label

```python
edited_y_label = []

for label in y_labels:
    if label[-3:] == 'xml':
        continue
    arr = label.split('_')
    edited_y_label.append(arr[0])
```

```python
model = tf.keras.Sequential()

model.add(tf.keras.layers.Conv2D(filters=32,
    kernel_size=(3, 3),
    activation='relu',
    input_shape=(64, 64, 3)))
model.add(tf.keras.layers.Conv2D(filters=32,
    kernel_size=(5, 5),
    activation='relu',
    input_shape=(64, 64, 3)))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(units=2, activation='softmax'))
```
